Llamar una BD para usar de ejemplo

```python
# Importa las bibliotecas necesarias
from sklearn import datasets
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

Para hacer un gráfico de dispersión con matplotlib utilizando el conjunto de datos digits, primero necesitarás reducir la dimensionalidad de los datos, ya que cada imagen en digits es una matriz de 8x8 (64 dimensiones) y un gráfico de dispersión es 2D. Una forma común de hacer esto es usando PCA (Principal Component Analysis).

```python
# Carga el conjunto de datos digits
digits = datasets.load_digits()

# Realiza PCA para reducir a 2 dimensiones
pca = PCA(n_components=2)
projected = pca.fit_transform(digits.data)
```

```python
fig, ax = plt.subplots(figsize=(7, 3), tight_layout=True)

ax.scatter(x=projected[:, 0], # Posición horizontal de los puntos
           y=projected[:, 1], # Posición vertical de los puntos
           marker='o', # forma de los puntos
           s=20, # Tamaño de los puntos
           c='r', # Color de los puntos
           alpha=1.) # Opacidad de los puntos

ax.set_xlabel('Componente 1')
ax.set_ylabel('Componente 2');
```

```python
fig, ax = plt.subplots(figsize=(7, 4), tight_layout=True, facecolor='w')

ax.hist(projected[1:] - projected[:-1], # Datos
        bins=20, # Cantidad de cajones por columna
        density=False, # Normalización
        histtype='bar', color=None); # Aspectos visuales

ax.set_xlabel('Componentes');
```

```python
# Crea el gráfico de dispersión
plt.scatter(projected[:, 0], projected[:, 1],
            c=digits.target, edgecolor='none', alpha=0.5,
            cmap=plt.cm.get_cmap('ocean', 10))
plt.xlabel('Componente 1')
plt.ylabel('Componente 2')
plt.colorbar()

# Muestra el gráfico
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(7, 4), tight_layout=True, facecolor='w')
ax.grid()
ax.stackplot(range(len(projected)), # Esto corresponde al eje horizontal
             projected[:, 0], # Los siguientes datos se graficaran en el eje vertical
             projected[:, 1],
             alpha=0.5, labels=('Componente 1', 'Componente 2'));
plt.legend(loc=2);
```

```python
# Crear un histograma para una columna específica (por ejemplo, la primera columna)
plt.hist(digits.data[:, 0], bins=10, edgecolor='black')
plt.title('Histograma de la primera columna en el conjunto de datos "digits"')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.grid(True)  # Agregar una cuadrícula opcional
plt.show()
```
